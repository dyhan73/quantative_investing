```python
import sys
import os
```

```python
# 디렉터리의 파일 목록
# from os.path import isfile, join
fi_rep_path = '../data/financial_reports'
fi_subdirs = [f for f in os.listdir(fi_rep_path) if os.path.isdir(os.path.join(fi_rep_path, f))]
fi_subdirs.sort(reverse=True)
fi_subdirs
```

```python
financial_report_files = []
for sdir in fi_subdirs:
    new_path = fi_rep_path+'/'+sdir
    fi_reps = [f for f in os.listdir(new_path) if os.path.isfile(os.path.join(new_path, f))]
    fi_reps = [f for f in fi_reps if 'Eng' not in f and 'Rank' not in f and 'ENG' not in f and '영문' not in f and 'hwp' not in f and 'txt' not in f]
#     fi_reps = [f for f in fi_reps if '한글' in f or '국문' in f or 'kor' in f or 'KOR' in f or 'Kor' in f]
    
    for rep in fi_reps:
        financial_report_files.append('%s/%s'%(new_path, rep))
print(len(financial_report_files))
financial_report_files

```
