```python
import pandas as pd
import numpy as np
import re

pd.options.display.float_format = '{:,.0f}'.format
```

```python
reports = pd.read_excel('../data/financial_reports/2019년12월확정실적/201912(한글).xlsx', sheet_name='주재무제표', header=[1, 2])
columns = [t[0] if t[1].startswith('Unnamed') else "%s-%s"%(re.sub('\n', '', t[0]), t[1]) for t in reports.keys()]
reports.columns = columns

cols = ('시장', '종목코드', '회사명', '회계기준', '산업코드', '산업명', '자산총계', '유동자산', '자본총계', '부채총계',
        '매출액-', '매출원가', '매출총이익', '영업이익(보고서기재)', '당기순이익', '영업활동으로인한현금흐름', '부채비율')
reports = reports[[k for k in reports.keys() if k.startswith(cols) and '비교' not in k]]
reports['종목코드'] = reports['종목코드'].str[1:]
```
